```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import os

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16
).to("cuda")
```

```python
input_folder = "/content"  # Root folder where your uploaded PNGs are
output_folder = "/content/deformed_faces"
os.makedirs(output_folder, exist_ok=True)

deform_prompt = "the same person with slight burn marks "

# Loop through all PNGs in root directory
for filename in os.listdir(input_folder):
    if filename.endswith(".png"):
        print(f"Processing: {filename}")
        image_path = os.path.join(input_folder, filename)
        image = Image.open(image_path).convert("RGB").resize((512, 512))

        result = pipe(
            prompt=deform_prompt,
            image=image,
            strength=0.6,
            guidance_scale=7.5
        )

        result_image = result.images[0]
        result_image.save(os.path.join(output_folder, f"deformed_{filename}"))
```
